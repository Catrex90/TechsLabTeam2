```python
import pandas as pd

## matplotlib benötigen wir für die Visualisierung des Datensatzes
import csv
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/selinaycengiz/Documents/TechLabs Group 2/hotel_bookings.csv")
df.head()
```

```python
# Erstelle seperate DataFrames für beide Hoteltypen
resort_hotel_data = df[df['hotel'] == 'Resort Hotel']
city_hotel_data = df[df['hotel'] == 'City Hotel']
print(resort_hotel_data)
print(city_hotel_data)
```

```python
# Gäste im Resort Hotel nach Monat zählen und die Daten in einem neuen DataFrame speichern
resort_guests = resort_hotel_data['arrival_date_month'].value_counts().reset_index()
# Umbenennen
resort_guests.columns=['month','number of guests']
resort_guests
```

```python
#Ranken der Monate 
resort_guests_sorted = resort_guests.sort_values(by='number of guests', ascending=False)
print(resort_guests_sorted)
```

```python
# Gäste im City Hotel nach Monat zählen und die Daten in einem neuen DataFrame speichern
city_guests = city_hotel_data['arrival_date_month'].value_counts().reset_index()
# Umbenennen
city_guests.columns=['month','number of guests']
city_guests
```

```python
#Ranken der Monate 
city_guests_sorted = city_guests.sort_values(by='number of guests', ascending=False)
print(city_guests_sorted)
```

```python
# Diagramm erstellen
plt.figure(figsize=(12, 5))

# Daten plotten
plt.bar(city_guests_sorted['month'], city_guests_sorted['number of guests'], color='lightcoral')

# Diagramm-Titel und Achsentitel hinzufügen
plt.title('Anzahl der Gäste im City Hotel pro Monat')
plt.xlabel('Monat')
plt.ylabel('Anzahl der Gäste')

# Diagramm anzeigen
plt.show()
```

```python
# Diagramm erstellen
plt.figure(figsize=(12, 5))

# Daten plotten
plt.bar(resort_guests_sorted['month'], resort_guests_sorted['number of guests'], color='lightblue')

# Diagramm-Titel und Achsentitel hinzufügen
plt.title('Anzahl der Gäste im Resort Hotel pro Monat')
plt.xlabel('Monat')
plt.ylabel('Anzahl der Gäste')

# Diagramm anzeigen
plt.show()
```
