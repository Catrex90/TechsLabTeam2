```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Hotel inkl. Country Codes (hotel shape war: 119390 rows × 32 columns, jetzt plus 5 Spalten für Country Codes Merge)
df_hotel = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Edited/Hotel_CountryCodes.csv", encoding="ISO-8859-1")


#Spaltennamen umändern, damit sie zu Happiness passen
df_hotel.rename(columns={'arrival_date_year': 'Year', "country_common" : "Country"}, inplace=True)

display(df_hotel)
```

```python
df_2015 = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Raw/Happiness/World Happiness/2015.csv")
df_2015['Year'] = 2015  # Hinzufügen einer Spalte mit dem Jahr 2015
df_2016 = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Raw/Happiness/World Happiness/2016.csv")
df_2016['Year'] = 2016  # Hinzufügen einer Spalte mit dem Jahr 2016
df_2017 = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Raw/Happiness/World Happiness/2017_2.csv", sep=";" )
df_2017['Year'] = 2017  # Hinzufügen einer Spalte mit dem Jahr 2017

#print(df_2015.head())
#print(df_2016.head())
print(df_2017.head())
```

```python
# Anzeigen der vorhandenen Spaltennamen in df_2015
print("Spaltennamen in df_2015:", df_2015.columns)

# Anzeigen der vorhandenen Spaltennamen in df_2016
print("Spaltennamen in df_2016:", df_2016.columns)

# Anzeigen der vorhandenen Spaltennamen in df_2017
print("Spaltennamen in df_2017:", df_2017.columns)
```

```python
# Bereinigen der Spaltennamen von Leerzeichen am Anfang oder Ende
df_2015.columns = df_2015.columns.str.strip()
df_2016.columns = df_2016.columns.str.strip()
df_2017.columns = df_2017.columns.str.strip()
```

```python
# Entfernen der gewünschten Spalten

# Direktes Löschen der Spalte 
if 'Standard Error' in df_2015.columns:
    del df_2015['Standard Error']

# Anzeigen der DataFrame-Kopfzeilen nach Entfernung der Spalte
print(df_2015.head())


if 'Standard Error' in df_2015.columns:
    del df_2015['Standard Error']

# Anzeigen der DataFrame-Kopfzeilen nach Entfernung der Spalte
print(df_2015.head())


# Überprüfen und Entfernen der gewünschten Spalten in df_2016
columns_to_remove_2016 = ['Lower Confidence Interval', 'Upper Confidence Interval']
for column in columns_to_remove_2016:
    if column in df_2016.columns:
        df_2016 = df_2016.drop([column], axis=1)

# Überprüfen und Entfernen der gewünschten Spalten in df_2017
columns_to_remove_2017 = ['Whisker high', 'Whisker low']
for column in columns_to_remove_2017:
    if column in df_2017.columns:
        df_2017 = df_2017.drop([column], axis=1)

# Zusammenführen der DataFrames
df_happiness = pd.concat([df_2015, df_2016, df_2017], ignore_index=True)

print(df_happiness)

# Speichern als CSV-Datei
#df_happiness.to_csv('/Users/carareusch/Desktop/TechsLabTeam2/TechsLabTeam2-3/Data/Edited/Combined_Happiness.csv', index=False)
```

```python
df_happiness_and_hotel = pd.merge(df_hotel, df_happiness, on=["Country", "Year"], how="left")

#Data cleaning: Alle Buchungen die gecancelt wurden oder wo keine Gäste gekommen sind streichen.
df_happiness_and_hotel = df_happiness_and_hotel[df_happiness_and_hotel["is_canceled"]==0]
df_happiness_and_hotel["total_guests"]= df_happiness_and_hotel["adults"] + df_happiness_and_hotel["children"] + df_happiness_and_hotel["babies"]
df_happiness_and_hotel = df_happiness_and_hotel[df_happiness_and_hotel["total_guests"]!=0]

#display(df_happiness_and_hotel)

print(len(df_hotel))
print(len(df_happiness_and_hotel))

#df_happiness_and_hotel.to_csv('/Users/carareusch/Desktop/TechsLabTeam2/TechsLabTeam2-3/Data/Edited/Happiness_and_Hotel_Combined.csv', index=False)
```
