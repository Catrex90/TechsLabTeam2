```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Karte einlesen
shapefile = "../Data/Raw/Maps/ne_50m_admin_0_countries.shp"
#gdf = gpd.read_file(shapefile)

#gdf = gpd.read_file(shapefile)[
#    ["ADMIN", "ADM0_A3", "geometry", "POP_EST"]
#]

gdf = gpd.read_file(shapefile)[
    ["ADMIN", "ADM0_A3", "geometry"]
]

display(gdf)
#print(gdf.head())
#print(len(gdf.columns))

# Spaltennamen festlegen
gdf.columns = ["ADMIN", "country_code", "geometry"]

# Antarktis entfernen, abbr. ATA
gdf = gdf[gdf.ADMIN != "Antarctica"]

# Anzahl Reisen einlesen
traveldf = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Raw/Hotel/hotel_bookings.csv", encoding="ISO-8859-1")

display(traveldf)

#freq = df["country"].value_counts()
#print(freq)

# Frequenz an Ländern ohne is_canceled
freq_df = df[df["is_canceled"] == 0]["country"].value_counts().reset_index()
print(freq)

#freq_df = df["country"].value_counts().reset_index()
freq_df.columns = ["country", "value_counts"]
print(freq_df)
```

```python

# Frequenz Länder in Karte integrieren
merged = gdf.merge(
    freq_df, left_on="country_code", right_on="country", how="left"
)

print(merged)
```

```python

# Nicht genannte Länder auf 0 setzen
merged["value_counts"] = merged["value_counts"].fillna(0)

# Plotten
fig, ax = plt.subplots(1, figsize=(15, 15))
p = merged.plot(
    column="value_counts",
    legend=True,
    ax=ax,
    cmap="OrRd",
    scheme="FisherJenks",
    legend_kwds={
        "loc": "lower left",
        "title": "Counts of Travels from specific country",
    },
)

# Hintergrundfarbe setzen
p.set_facecolor("gray")

# Achsenbeschriftung entfernen
ax.set_yticklabels([])
ax.set_xticklabels([])

# Grafik speichern
plt.savefig("Output\Figures\TravelPerCountry_Map.jpg", bbox_inches="tight", dpi=300)
```

```python
# Anzeige sortiert nach max Länder

#print(merged[["ADMIN", "value_counts"]])

sorted_merged = merged.sort_values(by="value_counts", ascending=False)
#display(sorted_merged[["ADMIN", "value_counts"]])

print(sorted_merged.head(25)[["ADMIN", "value_counts"]])

```

```python
# Pie Chart most common countries

import matplotlib.pyplot as plt

labels = sorted_merged["ADMIN"]
sizes = sorted_merged["value_counts"]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

```python
# Pie Chart most common countries, head 16

import matplotlib.pyplot as plt

labels = sorted_merged.head(16)["ADMIN"]
sizes = sorted_merged.head(16)["value_counts"]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)

# Grafik speichern
plt.savefig("Output\Figures\TravelPerCountry_Pie.jpg", bbox_inches="tight", dpi=300)
```

```python
# histogram (bar plot), From where are most guests coming?

import matplotlib.pyplot as plt

#show just top 25
fig, ax = plt.subplots()
labels = sorted_merged.head(25)["ADMIN"]
sizes = sorted_merged.head(25)["value_counts"]

ax.bar(labels, sizes)

ax.set_ylabel('visits 2015/07/01 - 2017/08/31')
ax.set_title('From where are most guests coming (Top 25)? ')
#ax.legend(title='countries')

# shows labels vertical
plt.xticks(rotation='vertical')
plt.show()

# Grafik speichern
plt.savefig("Output\Figures\TravelPerCountry_Bar25.jpg", bbox_inches="tight", dpi=300)
```
