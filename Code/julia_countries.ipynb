```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Karte einlesen
shapefile = "../Data/Raw/Maps/ne_50m_admin_0_countries.shp"
#gdf = gpd.read_file(shapefile)

#gdf = gpd.read_file(shapefile)[
#    ["ADMIN", "ADM0_A3", "geometry", "POP_EST"]
#]

gdf = gpd.read_file(shapefile)[
    ["ADMIN", "ADM0_A3", "geometry"]
]

display(gdf)
#print(gdf.head())
#print(len(gdf.columns))

# Spaltennamen festlegen
gdf.columns = ["ADMIN", "country_code", "geometry"]

# Antarktis entfernen, abbr. ATA
gdf = gdf[gdf.ADMIN != "Antarctica"]

# Anzahl Reisen einlesen
traveldf = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Raw/Hotel/hotel_bookings.csv", encoding="ISO-8859-1")

display(traveldf)

#freq = df["country"].value_counts()
#print(freq)

# Frequenz an Ländern ohne is_canceled
freq_df = df[df["is_canceled"] == 0]["country"].value_counts().reset_index()
print(freq)

#freq_df = df["country"].value_counts().reset_index()
freq_df.columns = ["country", "value_counts"]
print(freq_df)
```

```python

# Frequenz Länder in Karte integrieren
merged = gdf.merge(
    freq_df, left_on="country_code", right_on="country", how="left"
)

print(merged)
```

```python

# Nicht genannte Länder auf 0 setzen
merged["value_counts"] = merged["value_counts"].fillna(0)

# Plotten
fig, ax = plt.subplots(1, figsize=(15, 15))
p = merged.plot(
    column="value_counts",
    legend=True,
    ax=ax,
    cmap="OrRd",
    scheme="FisherJenks",
    legend_kwds={
        "loc": "lower left",
        "title": "Counts of Travels from specific country",
    },
)

# Hintergrundfarbe setzen
p.set_facecolor("gray")

# Achsenbeschriftung entfernen
ax.set_yticklabels([])
ax.set_xticklabels([])

# Grafik speichern
plt.savefig("Output\Figures\TravelPerCountry_Map.jpg", bbox_inches="tight", dpi=300)
```

```python
# Anzeige sortiert nach max Länder

#print(merged[["ADMIN", "value_counts"]])

sorted_merged = merged.sort_values(by="value_counts", ascending=False)
#display(sorted_merged[["ADMIN", "value_counts"]])

print(sorted_merged.head(25)[["ADMIN", "value_counts"]])

```

```python
# Pie Chart most common countries

import matplotlib.pyplot as plt

labels = sorted_merged["ADMIN"]
sizes = sorted_merged["value_counts"]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

# Where most guests are come from, City vs Resort Hotel

```python
# imports
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Karte einlesen
shapefile = "../Data/Raw/Maps/ne_50m_admin_0_countries.shp"
#gdf = gpd.read_file(shapefile)

#gdf = gpd.read_file(shapefile)[
#    ["ADMIN", "ADM0_A3", "geometry", "POP_EST"]
#]

gdf = gpd.read_file(shapefile)[
    ["ADMIN", "ADM0_A3", "geometry"]
]

#display(gdf)
#print(gdf.head())
#print(len(gdf.columns))

# Spaltennamen festlegen
gdf.columns = ["ADMIN", "country_code", "geometry"]

# Antarktis entfernen, abbr. ATA
gdf = gdf[gdf.ADMIN != "Antarctica"]


```

```python
# Anzahl Reisen einlesen
df = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Raw/Hotel/hotel_bookings.csv", encoding="ISO-8859-1")

# Frequenz an Ländern ohne is_canceled (Resort vs City)
freq = df[(df["is_canceled"] == 0)]["country"].value_counts().reset_index()
freq_City = df[(df["is_canceled"] == 0) & (df["hotel"] == "City Hotel")]["country"].value_counts().reset_index()
freq_Resort = df[(df["is_canceled"] == 0) & (df["hotel"] == "Resort Hotel")]["country"].value_counts().reset_index()

#print(freq)
#print(freq_City)
#print(freq_Resort)

#freq_df = df["country"].value_counts().reset_index()
freq_City.columns = ["country", "City_value_counts"]
display(freq_City)

freq_Resort.columns = ["country", "Resort_value_counts"]
display(freq_Resort)
```

```python
# Merge gdf and hotel dfs
# Frequenz Länder in Karte integrieren
merged_City = gdf.merge(
    freq_City, left_on="country_code", right_on="country", how="left"
)

#display(merged_City)

merged_all = merged_City.merge(
    freq_Resort, left_on="country_code", right_on="country", how="left"
)

display(merged_all)

# Nicht genannte Länder auf 0 setzen
merged_all["City_value_counts"] = merged_all["City_value_counts"].fillna(0)
merged_all["Resort_value_counts"] = merged_all["Resort_value_counts"].fillna(0)

display(merged_all)
```

```python
# Plotten City
fig, ax = plt.subplots(1, figsize=(15, 15))
p = merged_all.plot(
    column="City_value_counts",
    legend=True,
    ax=ax,
    cmap="OrRd",
    scheme="FisherJenks",
    legend_kwds={
        "loc": "lower left",
        "title": "Counts of Travels from specific country",
    },
)

# Hintergrundfarbe setzen
p.set_facecolor("gray")

# Achsenbeschriftung entfernen
ax.set_yticklabels([])
ax.set_xticklabels([])

# Grafik speichern
plt.savefig("Output\Figures\TravelPerCountry_Map_City.jpg", bbox_inches="tight", dpi=300)
```

```python
# Plotten Resort
fig, ax = plt.subplots(1, figsize=(15, 15))
p = merged_all.plot(
    column="Resort_value_counts",
    legend=True,
    ax=ax,
    cmap="OrRd",
    scheme="FisherJenks",
    legend_kwds={
        "loc": "lower left",
        "title": "Counts of Travels from specific country",
    },
)

# Hintergrundfarbe setzen
p.set_facecolor("gray")

# Achsenbeschriftung entfernen
ax.set_yticklabels([])
ax.set_xticklabels([])

# Grafik speichern
plt.savefig("Output\Figures\TravelPerCountry_Map_Resort.jpg", bbox_inches="tight", dpi=300)
```

```python
# Anzeige sortiert nach max Länder

#print(merged[["ADMIN", "value_counts"]])

sorted_merged_City = merged_all.sort_values(by="City_value_counts", ascending=False)
#display(sorted_merged[["ADMIN", "value_counts"]])

display(sorted_merged_City.head(25)[["ADMIN", "City_value_counts"]])

sorted_merged_Resort = merged_all.sort_values(by="Resort_value_counts", ascending=False)
#display(sorted_merged[["ADMIN", "value_counts"]])

display(sorted_merged_Resort.head(25)[["ADMIN", "Resort_value_counts"]])
```

```python
import matplotlib.pyplot as plt

# Plot for City
labels_city = sorted_merged_City.head(16)["ADMIN"]
sizes_city = sorted_merged_City.head(16)["City_value_counts"]

fig, ax = plt.subplots()
ax.pie(sizes_city, labels=labels_city)
ax.set_title('City Distribution')
plt.show()

# Plot for Resort
labels_resort = sorted_merged_Resort.head(16)["ADMIN"]
sizes_resort = sorted_merged_Resort.head(16)["Resort_value_counts"]

fig, ax = plt.subplots()
ax.pie(sizes_resort, labels=labels_resort)
ax.set_title('Resort Distribution')
plt.show()
```

```python
# Pie Chart most common countries, head 14, Country Color Code the same

import matplotlib.pyplot as plt

# Farben je Land gleich halten in den beiden Pie Figures

# Alle einzigartigen Länder aus beiden Dataframes sammeln
all_countries = sorted_merged_City.head(16)["ADMIN"].unique()
all_countries = set(all_countries)  # Um sicherzustellen, dass nur einzigartige Länder berücksichtigt werden
all_countries.update(sorted_merged_Resort.head(16)["ADMIN"].unique())

# Farben generieren
colors = plt.cm.tab20.colors[:len(all_countries)]  # Verwenden Sie eine Farbpalette mit genügend Farben

# Farbzuordnung erstellen
color_map = dict(zip(all_countries, colors))


# plot city

labels_city = sorted_merged_City.head(14)["ADMIN"]
sizes_city = sorted_merged_City.head(14)["City_value_counts"]

fig, ax = plt.subplots()
colors_city = [color_map.get(country, "gray") for country in labels_city]
ax.pie(sizes_city, labels=labels_city, colors=colors_city)
ax.set_title("City Hotel")
plt.show

# Grafik speichern
plt.savefig("Output\Figures\TravelPerCountry_Pie_City.jpg", bbox_inches="tight", dpi=300)


# plot Resort

labels_resort = sorted_merged_Resort.head(14)["ADMIN"]
sizes_resort = sorted_merged_Resort.head(14)["Resort_value_counts"]

fig, ax = plt.subplots()
colors_resort = [color_map.get(country, "gray") for country in labels_resort]
ax.pie(sizes_resort, labels=labels_resort, colors=colors_resort)
ax.set_title("Resort Hotel")
plt.show

# Grafik speichern
plt.savefig("Output\Figures\TravelPerCountry_Pie_Resort.jpg", bbox_inches="tight", dpi=300)

```
