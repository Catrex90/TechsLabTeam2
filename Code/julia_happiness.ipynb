## Merge Hotel Data (with Country Codes) with Happiness Data 2015

```python
import pandas as pd


# Data load Happiness Score 2015

df_15 = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Raw/Happiness/World Happiness/2015.csv")

complete_columns = list(df_15.columns)
display(complete_columns)

# add column "Year" with the value "2015"

df_15['Year'] = int(2015)

complete_columns = list(df_15.columns)
display(complete_columns)

display(df_15)
```

```python
# Data load Happiness Score 2015

df_16 = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Raw/Happiness/World Happiness/2016.csv")

complete_columns = list(df_16.columns)
display(complete_columns)

# add column "Year" with the value "2015"

df_16['Year'] = int(2016)

display(df_16)
```

```python
## merge of hotel happiness 2015 and 2016
df_15_16 = pd.concat([df_15, df_16])

display(df_15_16)
```

```python
df_17 = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Raw/Happiness/World Happiness/2017.csv")

complete_columns = list(df_17.columns)
display(complete_columns)

# add column "Year" with the value "2017"

df_17['Year'] = int(2017)

display(df_17)
```

```python
# Merge 2015_2016 and 2017
df_15_16_17 = pd.concat([df_15_16, df_17])

display(df_15_16_17)

complete_columns = list(df_15_16_17.columns)
display(complete_columns)
```

```python
# Hotel inkl. Country Codes (hotel shape war: 119390 rows × 32 columns, jetzt plus 5 Spalten für Country Codes Merge)

df_travel = pd.read_csv("C:/Users/jkr/Documents/9 Computer/Informatik/Techlabs/DataScienceTrack/Project_DS_2/TechsLabTeam2/Data/Edited/Hotel_CountryCodes.csv", encoding="ISO-8859-1")
display(df_travel)

# df15_16_17 Happiness Daten

## merge of hotel (codes) and happiness data frame 
df_travel_happy = pd.merge(df_travel, df_15_16_17, left_on=["country_common", "arrival_date_year"], right_on=["Country", "Year"], how="left")
display(df_travel_happy)

complete_columns = list(df_travel_happy.columns)
display(complete_columns)
```
